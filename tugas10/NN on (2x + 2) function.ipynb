```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt

# Membuat data latih
X_train = np.linspace(0, 10, 100)  # Data input (x)
y_train = 2 * X_train + 2 + np.random.normal(0, 2, 100)  # Data output (2x + 2 + noise)

# Membuat model Neural Network
model = Sequential()
model.add(Dense(units=1, input_dim=1))  # Layer input dengan satu neuron

# Mengonfigurasi model untuk pelatihan
model.compile(optimizer='sgd', loss='mean_squared_error')

# Melatih model dengan data latih
model.fit(X_train, y_train, epochs=100, verbose=0)

# Membuat data uji
X_test = np.linspace(0, 10, 20)  # Data input untuk pengujian

# Menggunakan model untuk membuat prediksi
y_pred = model.predict(X_test)

# Plot hasil
plt.scatter(X_train, y_train, label='Data Latih')
plt.plot(X_test, y_pred, color='red', label='Prediksi')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Membuat model Neural Network dengan 3 layer
model = Sequential()
model.add(Dense(units=10, input_dim=1, activation='relu'))  # Layer input dengan 10 neuron
model.add(Dense(units=5, activation='relu'))  # Hidden layer dengan 5 neuron
model.add(Dense(units=1))  # Layer output dengan 1 neuron

# Mengonfigurasi model untuk pelatihan
model.compile(optimizer='sgd', loss='mean_squared_error')

# Melatih model dengan data latih
model.fit(X_train, y_train, epochs=100, verbose=0)

# Membuat data uji
X_test = np.linspace(0, 10, 20)  # Data input untuk pengujian

# Menggunakan model untuk membuat prediksi
y_pred = model.predict(X_test)

# Plot hasil
plt.scatter(X_train, y_train, label='Data Latih')
plt.plot(X_test, y_pred, color='red', label='Prediksi')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Mengonfigurasi model untuk pelatihan dengan SGD sebagai optimizer dan MSE sebagai fungsi loss
model.compile(optimizer='sgd', loss='mean_squared_error')

# Melatih model dengan data latih
history = model.fit(X_train, y_train, epochs=100, verbose=0)

# Membuat data uji
X_test = np.linspace(0, 10, 20)  # Data input untuk pengujian

# Menggunakan model untuk membuat prediksi
y_pred = model.predict(X_test)

# Plot hasil dan kurva pelatihan
plt.figure(figsize=(12, 5))

# Plot data latih dan prediksi
plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, label='Data Latih')
plt.plot(X_test, y_pred, color='red', label='Prediksi')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Prediksi Model')

# Plot kurva pelatihan
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title('Kurva Pelatihan')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')

plt.tight_layout()
plt.show()
```
